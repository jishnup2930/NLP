```python
article='In computing, stop words are words which are filtered out before or \
after processing of natural language data (text).[1] Though "stop words" usually \
refers to the most common words in a language, there is no single universal list of \
stop words used by all natural language processing tools, and indeed not all tools \
even use such a list. Some tools specifically avoid removing these stop words to \
support phrase search.'
```

```python
nlp=spacy.load('en_core_web_sm')#pre trained module for connecting words
```

```python
spacy_stopwords=spacy.lang.en.stop_words.STOP_WORDS
print('Number of stop words:',len(spacy_stopwords))
print('first ten stop words:',list(spacy_stopwords)[:20])
```

```python
doc=nlp(article)
tokens=[token.text for token in doc if not token.is_stop]# add words in token which are not in stop words
print('Original article: ',article)
print()
print(doc)
print()
print(tokens)  #docs contain 'which' but it is not added to tokens as it is a stop word
```

# stopword_using nltk: method 2

```python
import nltk
print('NLTK Version:%s'%(nltk.__version__))
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
nltk_stopwords=nltk.corpus.stopwords.words('english')
print('Number of stop words: ',len(nltk_stopwords))
print('First ten stop words: ',list(nltk_stopwords)[:10])
```

```python
print(nltk_stopwords)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
tokens=nltk.tokenize.word_tokenize(article)
tokens=[i for i in tokens if not i in nltk_stopwords]#if the word is in tokens and not in stopwords it is assigned to tokens
print(article)
print()
print(tokens)#which,are,before..are stop words
```

# customizing stop word

```python
new_stopwords=['computing','processing','language']
stpwrd=nltk.corpus.stopwords.words('english')
stpwrd.extend(new_stopwords)
print(stpwrd)
```

```python
# removing custom words

removing_custom_words=[words for words in stpwrd if not words in new_stopwords] # add words which are in token  and dont add if it is in token
print(removing_custom_words) # removed the words computing, processing and language as they are in stopwords
```

# Gram implementation

```python
#function to generate n-grams

def extract_ngrams(data,num):
    n_grams=ngrams(nltk.word_tokenize(data),num)
#     print(n_grams)
    return[" ".join(grams) for grams in n_grams]
```

```python
import nltk
from nltk.util import ngrams
```

```python
data='A class is a blueprint for the object'
```

```python
print('1 gram: ',extract_ngrams(data,1))
print('2 gram: ',extract_ngrams(data,2))
print('3 gram: ',extract_ngrams(data,3))
print('4 gram: ',extract_ngrams(data,4))
```
